Questão 1 (Criação e Filtragem) Com base no conceito de criação de DataFrames e indexação booleana vistos em aula , crie um DataFrame chamado df_vendas a partir de um dicionário contendo as seguintes listas:

Produto: ['Teclado', 'Mouse', 'Monitor', 'Cadeira', 'CPU', 'Mousepad']

Valor: [150, 45, 1200, 850, 2500, 30]

Quantidade: [20, 35, 10, 5, 8, 50] Em seguida, filtre e exiba apenas as linhas onde o Valor do produto seja maior que 100.

```python
import pandas as pd

dados = {'produto': ['Teclado', 'Mouse', 'Monitor', 'Cadeira', 'CPU', 'Mousepad'],
         'valor': [150, 45, 1200, 850, 2500, 30],
         'quantidade': [20, 35, 10, 5, 8, 50]}

df_vendas = pd.DataFrame(dados)

print('DataFrame original:')
print(df_vendas)
print('-' * 30)
print('\nDataFrame filtrado (Valor > 100):')
df_vendas_filtrado = df_vendas[df_vendas['valor'] > 100]
print(df_vendas_filtrado)
```

Questão 2 (Estatísticas Descritivas e Limpeza) Utilizando o DataFrame df_vendas criado na questão anterior:

Exiba as estatísticas descritivas (média, desvio padrão, min, max, etc.) da coluna Valor .

Imagine que inserimos um dado faltante propositalmente. Crie uma nova linha com um valor nulo (None ou np.nan) na coluna Quantidade e, em seguida, utilize o método apropriado visto em aula para verificar a existência desse valor nulo no DataFrame .

Questão 5 (Subplots) Vamos praticar a criação de múltiplas visualizações em uma única figura . Crie uma figura (fig) que contenha dois subplots (pode ser 1 linha e 2 colunas, ou 2 linhas e 1 coluna):

No primeiro subplot, desenhe uma linha tracejada (--) .

No segundo subplot, desenhe uma linha pontilhada (:).

Os dados podem ser sequências numéricas simples (ex: x = [1, 2, 3], y = [1, 2, 3]).

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 2, 4, 7, 11]
y2 = [11, 7, 4, 2, 1]

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].plot(x, y1, linestyle='--', color='blue', label='Linha Tracejada')
axs[0].set_title('Primeiro Subplot (Tracejado)')
axs[0].set_xlabel('Eixo X')
axs[0].set_ylabel('Eixo Y')
axs[0].grid(True)
axs[0].legend()

axs[1].plot(x, y2, linestyle=':', color='red', label='Linha Pontilhada')
axs[1].set_title('Segundo Subplot (Pontilhado)')
axs[1].set_xlabel('Eixo X')
axs[1].set_ylabel('Eixo Y')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

Questão 6 (Estilização de Gráficos de Dispersão) Crie um gráfico de dispersão (scatter plot) com os seguintes dados: x = [1, 2, 3, 4, 5] e y = [2, 3, 5, 7, 11]. Para fixar a estilização , configure o gráfico para ter:

Marcadores no formato de triângulo (^ ou v).

Cor dos marcadores: vermelha (red ou 'r').

Título do gráfico: "Distribuição de Primos".

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.figure(figsize=(8, 6))
plt.scatter(x, y, marker='^', color='r', s=100)
plt.title('Distribuição de Primos')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.grid(True)
plt.show()
```

Questão 7 (Boxplot) Desafio de visualização estatística: O histograma nos mostra a distribuição, mas existe um gráfico excelente para visualizar a mediana e identificar outliers (valores discrepantes). Pesquise sobre o comando plt.boxplot() e gere um Boxplot simples utilizando uma lista de valores aleatórios (ex: dados = np.random.normal(100, 10, 200)).

```python
import matplotlib.pyplot as plt
import numpy as np

dados = np.random.normal(100, 10, 200)

plt.figure(figsize=(8, 6))
plt.boxplot(dados)
plt.title('Boxplot de Valores Aleatórios')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```

Questão 8 (Histograma com Curva Normal) Reproduza o exercício de análise estatística visto em aula .

Gere 1000 dados aleatórios com distribuição normal.

Plote o histograma desses dados.

Adicione a linha da curva normal teórica sobreposta ao histograma para comparar a distribuição dos dados gerados.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

mu, sigma = 100, 15
dados = np.random.normal(mu, sigma, 1000)

plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(dados, bins=30, density=True, alpha=0.6, color='g', label='Histograma dos Dados')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2, label='Curva Normal Teórica')

plt.title('Histograma com Curva Normal Sobreposta')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```
