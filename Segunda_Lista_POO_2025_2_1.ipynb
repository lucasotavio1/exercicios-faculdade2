```python
import re

# 1. Classe Livro
class Livro:
    def __init__(self, titulo, autor, isbn):
        self.titulo = titulo
        self.autor = autor
        self.isbn = isbn

    @classmethod
    def from_dict(cls, dados):
        """Construtor de classe que cria uma instância a partir de um dicionário."""
        return cls(dados['titulo'], dados['autor'], dados['isbn'])

    @staticmethod
    def validar_isbn(isbn):
        """Método estático para validar se o ISBN é uma string de 13 dígitos."""
        return re.match(r'^\d{13}$', isbn) is not None

# 2. Classe Membro
class Membro:
    def __init__(self, nome, id_membro):
        self.nome = nome
        self.id = id_membro
        self.livros_emprestados = []

# 3. Classe Biblioteca
class Biblioteca:
    def __init__(self):
        self.catalogo = {}
        self.membros = {}

    def adicionar_livro(self, livro):
        """Adiciona um livro ao catálogo após validar o ISBN."""
        if Livro.validar_isbn(livro.isbn):
            self.catalogo[livro.isbn] = livro
            print(f"Livro '{livro.titulo}' adicionado ao catálogo.")
        else:
            print(f"Erro: ISBN '{livro.isbn}' é inválido. Livro não adicionado.")

    def adicionar_membro(self, membro):
        """Adiciona um novo membro à biblioteca."""
        self.membros[membro.id] = membro
        print(f"Membro '{membro.nome}' com ID '{membro.id}' registrado.")

    def emprestar_livro(self, isbn, id_membro):
        """Empresta um livro a um membro, se ambos existirem."""
        if isbn not in self.catalogo:
            print(f"Erro: Livro com ISBN '{isbn}' não encontrado no catálogo.")
            return

        if id_membro not in self.membros:
            print(f"Erro: Membro com ID '{id_membro}' não encontrado.")
            return

        livro = self.catalogo.pop(isbn)
        membro = self.membros[id_membro]
        membro.livros_emprestados.append(livro)
        print(f"Livro '{livro.titulo}' emprestado para {membro.nome}.")

    def devolver_livro(self, isbn, id_membro):
        """Permite que um membro devolva um livro."""
        if id_membro not in self.membros:
            print(f"Erro: Membro com ID '{id_membro}' não encontrado.")
            return

        membro = self.membros[id_membro]
        livro_encontrado = None
        for livro in membro.livros_emprestados:
            if livro.isbn == isbn:
                livro_encontrado = livro
                break

        if livro_encontrado:
            membro.livros_emprestados.remove(livro_encontrado)
            self.catalogo[livro_encontrado.isbn] = livro_encontrado
            print(f"Livro '{livro_encontrado.titulo}' devolvido por {membro.nome}.")
        else:
            print(f"Erro: Livro com ISBN '{isbn}' não foi encontrado na lista de empréstimos do membro '{membro.nome}'.")

# --- Código de Teste ---
print("--- Inicializando o sistema ---")

# Criando instâncias de Livro usando o método de classe
livro1_data = {'titulo': 'O Hobbit', 'autor': 'J.R.R. Tolkien', 'isbn': '1234567890123'}
livro1 = Livro.from_dict(livro1_data)

livro2_data = {'titulo': '1984', 'autor': 'George Orwell', 'isbn': '9876543210987'}
livro2 = Livro.from_dict(livro2_data)

# Criando uma instância de Membro
membro1 = Membro("João", "M001")

# Criando a instância da Biblioteca
biblioteca = Biblioteca()

print("\n--- Adicionando livros e membros ---")
biblioteca.adicionar_livro(livro1)
biblioteca.adicionar_livro(livro2)
biblioteca.adicionar_membro(membro1)

print("\n--- Testando o empréstimo ---")
biblioteca.emprestar_livro(livro1.isbn, membro1.id)

print("\n--- Testando a devolução ---")
biblioteca.devolver_livro(livro1.isbn, membro1.id)
```

```python
import re

# 1. Classe Livro
# Usa __slots__ para otimizar a memória
class Livro:
    __slots__ = ['__titulo', '__autor', '__isbn']

    def __init__(self, titulo, autor, isbn):
        self.__titulo = titulo
        self.__autor = autor
        self.__isbn = isbn

    @property
    def titulo(self):
        """Getter para o título do livro."""
        return self.__titulo

    @property
    def autor(self):
        """Getter para o autor do livro."""
        return self.__autor

    @property
    def isbn(self):
        """Getter para o ISBN do livro."""
        return self.__isbn

    @classmethod
    def from_dict(cls, dados):
        """Construtor de classe para criar um Livro a partir de um dicionário."""
        return cls(dados['titulo'], dados['autor'], dados['isbn'])

    @staticmethod
    def validar_isbn(isbn):
        """Método estático para validar se o ISBN é uma string de 13 dígitos."""
        return re.match(r'^\d{13}$', isbn) is not None

# 2. Classe Membro
class Membro:
    def __init__(self, nome, id_membro):
        self.__nome = nome
        self.__id = id_membro
        self.__livros_emprestados = []

    @property
    def nome(self):
        """Getter para o nome do membro."""
        return self.__nome

    @property
    def id(self):
        """Getter para o ID do membro."""
        return self.__id

    @property
    def livros_emprestados(self):
        """Getter para a lista de livros emprestados."""
        return self.__livros_emprestados

# 3. Classe Biblioteca
class Biblioteca:
    def __init__(self):
        self.__catalogo = {}
        self.__membros = {}

    @property
    def catalogo(self):
        return self.__catalogo

    @property
    def membros(self):
        return self.__membros

    def adicionar_livro(self, livro):
        """Adiciona um livro ao catálogo após validar o ISBN."""
        # Note que usamos 'livro.isbn' que, por baixo dos panos,
        # chama o método getter do @property na classe Livro.
        if Livro.validar_isbn(livro.isbn):
            self.catalogo[livro.isbn] = livro
            print(f"Livro '{livro.titulo}' adicionado ao catálogo.")
        else:
            print(f"Erro: ISBN '{livro.isbn}' é inválido. Livro não adicionado.")

    def adicionar_membro(self, membro):
        """Adiciona um novo membro à biblioteca."""
        self.membros[membro.id] = membro
        print(f"Membro '{membro.nome}' com ID '{membro.id}' registrado.")

    def emprestar_livro(self, isbn, id_membro):
        """Empresta um livro a um membro, se ambos existirem."""
        if isbn not in self.catalogo:
            print(f"Erro: Livro com ISBN '{isbn}' não encontrado no catálogo.")
            return

        if id_membro not in self.membros:
            print(f"Erro: Membro com ID '{id_membro}' não encontrado.")
            return

        livro = self.catalogo.pop(isbn)
        membro = self.membros[id_membro]
        membro.livros_emprestados.append(livro)
        print(f"Livro '{livro.titulo}' emprestado para {membro.nome}.")

    def devolver_livro(self, isbn, id_membro):
        """Permite que um membro devolva um livro."""
        if id_membro not in self.membros:
            print(f"Erro: Membro com ID '{id_membro}' não encontrado.")
            return

        membro = self.membros[id_membro]
        livro_encontrado = None
        for livro in membro.livros_emprestados:
            if livro.isbn == isbn:
                livro_encontrado = livro
                break

        if livro_encontrado:
            membro.livros_emprestados.remove(livro_encontrado)
            self.catalogo[livro_encontrado.isbn] = livro_encontrado
            print(f"Livro '{livro_encontrado.titulo}' devolvido por {membro.nome}.")
        else:
            print(f"Erro: Livro com ISBN '{isbn}' não foi encontrado na lista de empréstimos do membro '{membro.nome}'.")

# --- Código de Teste ---
print("--- Inicializando o sistema ---")

# Criando instâncias de Livro usando o método de classe
livro1_data = {'titulo': 'O Hobbit', 'autor': 'J.R.R. Tolkien', 'isbn': '1234567890123'}
livro1 = Livro.from_dict(livro1_data)

livro2_data = {'titulo': '1984', 'autor': 'George Orwell', 'isbn': '9876543210987'}
livro2 = Livro.from_dict(livro2_data)

# Criando uma instância de Membro
membro1 = Membro("João", "M001")

# Criando a instância da Biblioteca
biblioteca = Biblioteca()

print("\n--- Adicionando livros e membros ---")
biblioteca.adicionar_livro(livro1)
biblioteca.adicionar_livro(livro2)
biblioteca.adicionar_membro(membro1)

print("\n--- Testando o empréstimo ---")
biblioteca.emprestar_livro(livro1.isbn, membro1.id)

print("\n--- Testando a devolução ---")
biblioteca.devolver_livro(livro1.isbn, membro1.id)
```
