Crie um array NumPy chamado vetor_A que contenha todos os números inteiros de 10 a 50 (inclusivo).

Use a função reshape para transformar esse vetor em uma matriz 5x8 chamada matriz_A.

Verifique o shape e o dtype da matriz_A.

Crie uma nova matriz matriz_A_float convertendo o tipo de dado (dtype) da matriz_A para float64  usando o método .astype().

Imprima a matriz_A_float, seu shape e seu dtype.

```python
import numpy as np


vetor_a = np.arange(10, 50)
matriz_a = np.reshape(vetor_a, (5, 8))

print(matriz_a)
print('-' * 30)
print(matriz_a.shape)
print('-' * 30)
print(matriz_a.dtype)
print('-' * 30)

matriz_a_float = matriz_a.astype(np.float64)
print(matriz_a_float)
```

Crie uma matriz 4x6 chamada matriz_E com dados aleatórios seguindo uma distribuição normal (use np.random.randn(4, 6)).

Calcule e imprima:

A média (mean) de toda a matriz.

O valor máximo (max) de toda a matriz.

A soma (sum) de cada coluna (use o parâmetro axis).

O desvio padrão (std) de cada linha (use o parâmetro axis).

```python
matriz_e = np.random.randn(4, 6)

print(matriz_e)
print('-' * 30)
print(np.mean(matriz_e))
print('-' * 30)
print(np.max(matriz_e))
print('-' * 30)
print(np.sum(matriz_e, axis=0))
print('-' * 30)
print(np.std(matriz_e, axis=1))
```

Crie a seguinte matriz 2x2: A = np.array([[5, 1], [3, 2]])

Pesquise e utilize a função np.linalg.det() para calcular e imprimir o determinante da matriz A.

Pesquise e utilize a função np.linalg.inv() para calcular e imprimir a matriz inversa de A.

Calcule o produto matricial (use np.dot())  entre A e sua inversa. (O que você espera que seja o resultado?)

```python
arr3 = np.array([[5, 1], [3, 2]])
matriz_f = np.reshape(arr3, (2, 2))

print(matriz_f)
print('-' * 30)
print(np.linalg.det(matriz_f))
print('-' * 30)
print(np.linalg.inv(matriz_f))
```
