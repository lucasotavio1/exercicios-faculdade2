3 - Implemente o Diagrama de Classes a seguir no Python. Instancie um objeto de cada classe e apresente a área de cada triângulo.

UML3.png

```python
import math


class Triangulo:
    def __init__(self, lado_a: float, lado_b: float, lado_c: float):
        self.lado_a = lado_a
        self.lado_b = lado_b
        self.lado_c = lado_c


class TrianguloEquilatero(Triangulo):
    def __init__(self, lado_a: float):
        super().__init__(lado_a, lado_a, lado_a)

    def area(self) -> float:
        return (self.lado_a ** 2 * math.sqrt(3)) / 4


class TrianguloIsoceles(Triangulo):
    def __init__(self, lado_a: float, lado_b: float):
        super().__init__(lado_a, lado_a, lado_b)

    def area(self) -> float:
        altura = math.sqrt(self.lado_a**2 - (self.lado_c / 2)**2)
        return (self.lado_c * altura) / 2


class TrianguloEscaleno(Triangulo):
    def __init__(self, lado_a: float, lado_b: float, lado_c: float):
        super().__init__(lado_a, lado_b, lado_c)

    def area(self) -> float:
        semi_perimetro = (self.lado_a + self.lado_b + self.lado_c) / 2
        return math.sqrt(semi_perimetro * (semi_perimetro - self.lado_a) * (semi_perimetro - self.lado_b) * (semi_perimetro - self.lado_c))


tri_equilatero = TrianguloEquilatero(10)
print(f'Área do triângulo equilátero: {tri_equilatero.area():.2f}')

triangulo_isoceles = TrianguloIsoceles(15, 12)
print(f'Área do triângulo isósceles: {triangulo_isoceles.area():.2f}')

triangulo_escaleno = TrianguloEscaleno(5, 6, 7)
print(f'Área do triângulo escaleno: {triangulo_escaleno.area():.2f}')
```
